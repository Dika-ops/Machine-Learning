### 1. **Analisis Data dan Perannya dalam Proses Pengambilan Keputusan Berbasis Data**

Analisis data adalah proses pengumpulan, pembersihan, pemrosesan, dan interpretasi data untuk mengungkapkan pola, hubungan, dan wawasan yang berguna. Pengambilan keputusan berbasis data bergantung pada fakta dan angka, yang memungkinkan tim untuk membuat keputusan yang lebih akurat daripada sekadar mengandalkan intuisi.

### 2. **Perbedaan Descriptive Analysis dan Exploratory Data Analysis (EDA)**

**Descriptive Analysis** adalah teknik analisis yang berfokus pada menyajikan informasi dasar dan ringkasan statistik dari data yang ada. Tujuannya adalah untuk memberikan gambaran umum mengenai data dengan menggunakan statistik deskriptif, seperti mean, median, mode, standar deviasi, dan sebagainya.

**Exploratory Data Analysis (EDA)** adalah pendekatan yang lebih mendalam dalam memahami data. Tujuannya adalah untuk mengeksplorasi data dengan tujuan mengidentifikasi pola, tren, hubungan antar variabel, dan potensi masalah dalam data. EDA melibatkan visualisasi data, seperti histogram.

### 3. **Pentingnya Data Cleaning dan Data Wrangling Sebelum Melakukan Analisis Lanjutan**

**Data Cleaning** adalah proses untuk menangani data yang hilang, duplikat, atau inkonsisten agar data menjadi lebih bersih dan siap untuk dianalisis.

**Data Wrangling** adalah proses pengolahan data untuk mengubah atau mengonversi data mentah menjadi format yang lebih mudah dianalisis.

Keduanya sangat penting karena data yang tidak bersih atau terstruktur dengan buruk dapat mengarah pada hasil yang tidak akurat, yang bisa mempengaruhi pengambilan keputusan.

### 4. **Peran Visualisasi Data dalam Menyampaikan Insight Secara Jelas**

Visualisasi data adalah cara untuk menyajikan data dalam bentuk grafik atau diagram yang mudah dipahami. Visualisasi data berfungsi untuk:

* Mempermudah pemahaman hubungan antar variabel.
* Menunjukkan distribusi data secara visual.
* Membantu dalam deteksi pola atau tren.
* Memudahkan pihak awam untuk mengerti untuk memahami data.

### 5. **Teknik Machine Learning yang Digunakan**
* **Regresi (Regression)**: Digunakan untuk memprediksi nilai kontinu, seperti memprediksi tingkat polusi berdasarkan faktor-faktor cuaca.
* **Klasifikasi (Classification)**: Digunakan untuk memprediksi kategori dari data, seperti mengklasifikasikan kualitas udara sebagai "baik" atau "buruk" berdasarkan konsentrasi polutan.
* **Clustering (Pengelompokan)**: Digunakan untuk mengelompokkan data ke dalam kategori yang memiliki kesamaan, seperti mengelompokkan waktu berdasarkan kondisi cuaca dan tingkat polusi yang serupa.
